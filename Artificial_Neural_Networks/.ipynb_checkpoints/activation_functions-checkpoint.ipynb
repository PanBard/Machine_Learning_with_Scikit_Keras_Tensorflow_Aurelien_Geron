```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Heaviside step function
def heaviside(x):
    return np.where(x >= 0, 1, 0)

# Create an array of input values
x = np.linspace(-10, 10, 400)

# Plotting
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Sigmoid plot
axs[0, 0].plot(x, sigmoid(x), label="Sigmoid")
axs[0, 0].set_title('Sigmoid Activation Function')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Tanh plot
axs[0, 1].plot(x, tanh(x), label="Tanh")
axs[0, 1].set_title('Tanh Activation Function')
axs[0, 1].grid(True)
axs[0, 1].legend()

# ReLU plot
axs[1, 0].plot(x, relu(x), label="ReLU")
axs[1, 0].set_title('ReLU Activation Function')
axs[1, 0].grid(True)
axs[1, 0].legend()

# Leaky ReLU plot
axs[1, 1].plot(x, leaky_relu(x), label="Leaky ReLU")
axs[1, 1].set_title('Leaky ReLU Activation Function')
axs[1, 1].grid(True)
axs[1, 1].legend()

# Heaviside step function plot
axs[2, 0].plot(x, heaviside(x), label="Heaviside Step Function", color='green')
axs[2, 0].set_title('Heaviside Step Function')
axs[2, 0].grid(True)
axs[2, 0].legend()

# Softmax plot (we need a range of values, so we create a 2D vector)
x2 = np.linspace(-2, 2, 100)
X2 = np.vstack([x2, x2[::-1]])  # Create a simple 2D input for softmax
softmax_vals = softmax(X2)

axs[2, 1].plot(x2, softmax_vals[0], label="Softmax (1st class)")
axs[2, 1].plot(x2, softmax_vals[1], label="Softmax (2nd class)", linestyle='dashed')
axs[2, 1].set_title('Softmax Activation Function')
axs[2, 1].grid(True)
axs[2, 1].legend()

plt.tight_layout()
plt.show()
```
