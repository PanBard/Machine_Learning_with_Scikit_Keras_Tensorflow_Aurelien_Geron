# Training a Binary Classifier
- predicting only two calss: if image is choosed_number or not 

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False)
print(mnist.keys() )
X, y = mnist.data, mnist.target
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
choosed_number = '5'

y_train_5 = (y_train == choosed_number)  # True for all 5s, False for all other digits
y_test_5 = (y_test == choosed_number)
```

```python
y_train_5
```

## RandomForestClassifier


```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(random_state=42)
```

```python
from sklearn.model_selection import cross_val_predict
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_5, cv=3,
                                    method="predict_proba")
```

```python
# image_index = 1 #  image index from train dataset
# print(f'prediction of image number {image_index+1}:  {y_probas_forest[image_index]}') # prediction

# import matplotlib.pyplot as plt
# image = X_train[image_index].reshape(28, 28)
# plt.figure(figsize=(2,2))
# plt.imshow(image, cmap="binary")
# plt.axis("off")
# plt.show()

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
for idx, image_data in enumerate(X[:10]):
    plt.subplot(10, 10, idx + 1)
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap="binary")
    plt.title(f"prediction: \n {y_probas_forest[idx]}", fontsize=8)
    plt.axis("off")
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
y_scores_forest = y_probas_forest[:, 1]
precisions_forest, recalls_forest, thresholds_forest = precision_recall_curve(y_train_5, y_scores_forest)
```

```python
plt.figure(figsize=(6, 5))  # extra code â€“ not needed, just formatting

plt.plot(recalls_forest, precisions_forest, "b-", linewidth=2,
         label="Random Forest")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.axis([0, 1, 0, 1])
plt.grid()
plt.legend(loc="lower left")

plt.show()
```
