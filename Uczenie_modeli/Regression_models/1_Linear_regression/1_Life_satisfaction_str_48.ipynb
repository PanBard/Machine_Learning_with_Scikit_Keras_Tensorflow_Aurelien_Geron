```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
# plt.show() # dont show yet, bc we want to show regression line 

# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Cyprus
X_new = [[30000]]  # new data for prediction
y_new = model.predict(X_new) # prediction 
print(f"for x= {X_new} y= {y_new}") # outputs [[6.30165767]]


# Plot the regression line
x_range = np.linspace(23_500, 62_500, 100).reshape(-1, 1)  # Range of GDP values
y_range = model.predict(x_range)  # Predicted life satisfaction for each GDP value
plt.plot(x_range, y_range, color='green', label='prediction line')  # Plot the regression line

# Show the predicted point on the plot
plt.scatter(X_new, y_new, color='red', label=f'Prediction for {X_new}', zorder=5)
plt.legend()

plt.show()
```
